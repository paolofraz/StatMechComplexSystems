### Task 1

```python
# SIS Model
u = 0.5

# Homogeneous (MF)
# lmbd = u/<k>
print("-- Homogeneous Case --")
lmbd1_c_homo = u/np.mean(list(dict(nx.degree(G1)).values()))
lmbd2_c_homo = u/np.mean(list(dict(nx.degree(G2)).values()))
print("Epidemic Threshold 1st Network =",lmbd1_c_homo)
print("Epidemic Threshold 2nd Network =",lmbd2_c_homo)

# Heterogeneous
# lmbd = u<k>/<k^2>
print("-- Heterogeneous Case --")
lmbd1_c_het = u*np.mean(list(dict(nx.degree(G1)).values()))/np.mean(np.array(list(dict(nx.degree(G1)).values()))**2)
lmbd2_c_het = u*np.mean(list(dict(nx.degree(G2)).values()))/np.mean(np.array(list(dict(nx.degree(G2)).values()))**2)
print("Epidemic Threshold 1st Network =",lmbd1_c_het)
print("Epidemic Threshold 2nd Network =",lmbd2_c_het)

# Quenched Mean-Field
# lmbd = u/max(eigenval(Adj))
print("-- Quenched Mean-Field Case --")
lmbd1_c_qmf = u/np.max(nx.adjacency_spectrum(G1)).real
lmbd2_c_qmf = u/np.max(nx.adjacency_spectrum(G2)).real
print("Epidemic Threshold 1st Network =",lmbd1_c_qmf)
print("Epidemic Threshold 2nd Network =",lmbd2_c_qmf)
```

```python
def Simulation(G1, provided_lambdas=False,lambdas=[],nlmbd=50, runs=50, percentage_infected=0.05,outfile=""):
    
    np.random.seed(2019)
    
    model = sis.SISModel(G1)
    print(model.get_status_map())

    if provided_lambdas == False:
        lambdas = np.linspace(0,1,nlmbd)
    else:
        lambdas = lambdas
        
    iterations = np.empty((len(lambdas),runs),dtype=int)

    for i, lmbd in enumerate(lambdas):
        # Model Configuration
        cfg = mc.Configuration()
        cfg.add_model_parameter('beta', lmbd)
        cfg.add_model_parameter('lambda', 0.5)
        cfg.add_model_parameter("percentage_infected", percentage_infected)
        model.set_initial_status(cfg)

        # Simulation execution
        for j in range(runs):
            it  = model.iteration_bunch(100)
            iterations[i,j] = it[-1]['node_count'][1]
        print(i,end=" ")
    
    if provided_lambdas == False:
        np.save("./iter"+str(G1)+outfile,iterations)
        print("File Saved as: "+"./iter"+str(G1)+outfile)
    else:
        np.save("./iter_c"+str(G1)+outfile,iterations)
        print("File Saved as: "+"./iter_c"+str(G1)+outfile)

def PlotPhaseDiag(G1,iterations,lambdas,runs=50,critical = False):
    data1 = np.array([np.repeat(lambdas,runs),iterations.flatten()/nx.number_of_nodes(G1)])
    df1   = pd.DataFrame(data=data1.T,columns=["Lambda","i [%]"])
    if critical == False:
        g = sns.relplot(x="Lambda", y="i [%]",kind="line", ci="sd", data=df1,aspect=1.2) #, ci="sd"
        g.set(title="Phase Diagram - "+str(G1))
        g.savefig("./PhaseDiagram"+str(G1)+".png",dpi=100)
        return g
        plt.show()
    else:
        g = sns.pointplot(x="Lambda", y="i [%]", data=df1,aspect=1.2, join=False, capsize=.4) #, ci="sd"
        g.set(title="Phase Diagram, Critical Region - "+str(G1))
        g.xaxis.set_major_formatter(FormatStrFormatter('%0.4f'))
        return g
```

```python
# definition of propensity rates
def a1(I,u=0.5):
    return u*I

def a2(S,I,lmbd):
    return (lmbd * I * S) / (S + I)

def a0(S,I,lmbd,u=0.5):
    return a1(I,u) + a2(S,I,lmbd)

# def of counting function
def count_SI(G):
    S = sum(value == 'S' for value in nx.get_node_attributes(G,'status').values()) 
    I = sum(value == 'I' for value in nx.get_node_attributes(G,'status').values())
    return S, I

# Gillespie
def Gillespie(S,I,lmbd):
    tau  = np.random.exponential(scale=a0(S,I,lmbd))
    aj   = np.ceil(np.random.rand()*a0(S,I,lmbd)) / a1(I)
    if aj <= 1:
        return 1, tau # first reaction occurs, I -> S
    else:
        return 2, tau # second reaction occurs, S+I -> 2I

def GetInfectedNodeIndex(G1,n=1):
    infected_dict = {k: v for k, v in G1.nodes().items() if v['status']=='I'}
    if bool(infected_dict) == False:
        raise ValueError('All the population is Susceptible!')
    return random.sample(infected_dict.keys(),n)

def FirstReaction(G,inode):
    G.nodes()[inode]['status'] = 'S' #recovered
    return G

def SecondReaction(G,inode):
    neighbors        = list(G1[inode])
    susceptible_dict = 
    
    G.nodes()[inode]['status'] = 'S' #recovered
    return G
```

```python
np.random.seed(2019)
lmbd = 0.1

Total_time = 10000
t = 0
T = 10000
S = np.empty(Total_time)
I = np.empty(Total_time)

nx.set_node_attributes(G1,'S',name='status')
starting_infected_percentage = 0.05
starting_infected_nodes      = np.random.choice(nx.number_of_nodes(G1),size=int(starting_infected_percentage*nx.number_of_nodes(G1)))
for x in starting_infected_nodes:
    G1.nodes[x]['status'] = 'I'
#G1.nodes(data='status')
S[0], I[0]    = count_SI(G1)
while t < T:
    reaction, tau = Gillespie(S[t],I[t],lmbd)
    infected_node = GetInfectedNodeIndex(G1)
    
    if reaction == 1:
        G1 = FirstReaction(G1,infected_node)
    elif reaction == 2:
        G1 = SecondReaction(G1,infected_node)
    else:
        print("Something wrong :(")
        break
    
    S[t+1:t+int(tau)], I[t+1:t+int(tau)] = count_SI(G1)
    t += int(tau)
```
