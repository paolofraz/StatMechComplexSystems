### Task 4

```python
def Constraints(sigma, axis=[0,1]):
    return np.mean(sigma, axis=axis[0]) , np.mean(np.sum(sigma, axis=axis[1])**2)/sigma.shape[axis[1]] 

def Hamiltonian(sigma, lmbd, k, S=299):
    return -np.sum(lmbd*sigma)-(k/S)*(np.sum(sigma)**2)

def Metropolis_old(lmbd,k,S=299, N = 10000):
    """
    Run Metropolis algorithm by starting from one random configuration and storing the last 20% with minimum energy
    """
    n     = N/5 # store last 20% 
    s     = np.random.choice([-1,1],S)
    E     = Hamiltonian(s,lmbd,k) # starting energy
    s_sum    = np.sum(s)
    #energies = np.zeros(int(N)) 
    s_config = np.zeros((int(N/5),S)) # store last 20% configurations
    for t in range(N):
        #energies[t] = E
        j        = np.random.choice(np.arange(S)) # pick a spin
        dE       = 2*lmbd[j]*s[j] +(4*k/S)*s[j]*(s_sum-s[j])
        if(dE < 0):
            s[j]=-s[j]
            E += dE
            s_sum += 2*s[j]
        else:
            p_acc = np.exp(-dE)
            if(np.random.uniform()<=p_acc):
                s[j]=-s[j]
                E += dE
                s_sum += 2*s[j]      
        if t >= N-n:
            s_config[int(t-N+n),:] = s
            #energies[int(t-N+n)] = E # store last 20% energies
    return np.array(s_config) , energies

def Metropolis(lmbd,k,nconifig=20,S=299, N = 2000):
    """
    Run Metropolis algorithm by starting from 20 random configuration and storing the last one with minimum energy
    """
    s        = np.random.choice([-1,1],size=(nconifig,S))
    energies = np.empty(nconifig) 
    s_config = np.empty((nconifig,S))
    for n in range(nconifig):
        E        = Hamiltonian(s[n,:],lmbd,k) # starting energy
        s_sum    = np.sum(s[n,:])
        for t in range(N):
            j        = np.random.choice(np.arange(S)) # pick a spin
            dE       = 2*lmbd[j]*s[n,j] +(4*k/S)*s[n,j]*(s_sum-s[n,j])
            if(dE < 0):
                s[n,j]=-s[n,j]
                E += dE
                s_sum += 2*s[n,j]
            else:
                p_acc = np.exp(-dE)
                if(np.random.uniform()<=p_acc):
                    s[n,j]=-s[n,j]
                    E += dE
                    s_sum += 2*s[n,j]
        s_config[n,:] = s[n,:]
        energies[n]   = E 
    return s_config, energies
```

```python
np.random.seed(2019)
# First Approach
runs = 1000

lmbd   = np.empty((runs,S))
 # use previous values as starting points
lmbd_old = np.arctanh(m)
pd.set_option('use_inf_as_na', True) # remove inf values
lmbd_old = lmbd_old.fillna(np.random.rand())
lmbd[0,:] = lmbd_old

k      = np.empty(runs)
k[0]   = np.random.randn(1)

m_trg, m2_trg = Constraints(sigma) # target values
eta    = 0.0001

for t in range(runs):
    s, energies       = Metropolis(lmbd[t],k[t])
    m_const, m2_const = Constraints(s)
    
    if t < runs-1:
        lmbd[t+1,:] = lmbd[t,:] + eta * (m_trg - m_const)
        k[t+1]      = k[t] + eta * (m2_trg - m2_const)
```

### Task 6

```python
def GaussianApprox(x):
    C    = np.cov(x, rowvar=False)
    M    = np.linalg.inv(C)
    lmbd = -np.matmul(M, np.mean(x, axis=0))
    return C, M,lmbd

x          = abundances.unstack().to_numpy()
x_mean     = abundances.mean(level='spID').to_numpy()
x_sdt      = abundances.std(level= 'spID').to_numpy()
most_abundant = x_mean > x_sdt

Cov, M, lmbd_gauss = GaussianApprox(x[:,most_abundant])

np.fill_diagonal(M,0) # set diagonal to 0
np.fill_diagonal(Cov,0) # set diagonal to 0
```

```python
fig, ax = plt.subplots(1,2,figsize = (10,6))
mat = ax[0].imshow(M,cmap='viridis')
plt.colorbar(mat,ax=ax[0])
ax[0].set_title("Interaction Matrix")
cov = ax[1].imshow(Cov,cmap='viridis')
plt.colorbar(cov,ax=ax[1])
ax[1].set_title("Covariance Matrix")
plt.show()
```

```python
sns.distplot(lmbd_gauss,rug=True,axlabel='$\mu_j$')
plt.ylabel('Density')
plt.title("Lagrangian Parameters for the Gaussian Approximation")
plt.savefig('./Images/Lambda_GaussApprox.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
import networkx as nx

G_M = nx.from_numpy_matrix(M)
nx.draw(G_M,alpha=0.5)
print(nx.info(G_M))
```

```python
W          = M.copy()
thetas     = np.linspace(0,0.2, num = 2000)
conn_comps = np.empty(thetas.size)
thresholds = np.zeros(thetas.size)

for i,theta in enumerate(thetas):
    W_star = W
    mask   = np.logical_and(-theta < W, W < theta)
    W_star[mask] = 0
    G = nx.from_numpy_matrix(W_star)
    conn_comps[i] = nx.number_connected_components(G)
    thresholds[i] = theta

theta_star = thresholds[np.max(np.argwhere(conn_comps==1))]
print("theta_star = ",np.min(theta_star))
plt.plot(thetas,conn_comps,'-')
plt.ylabel("Number of Connected Components")
plt.xlabel("Threshold $θ$")
plt.savefig('./Images/Thetas.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
W_star = M.copy()
mask   = np.logical_and(-theta_star < W_star, W_star < theta_star)
W_star[mask] = 0
G = nx.from_numpy_matrix(W_star)

interactions = np.unique(W_star.flatten())
sns.distplot(interactions,axlabel='$W_{ij}^{*}$')
plt.xlim((-0.1,0.1))
plt.ylabel('Density')
plt.title("Distribution of $W_{ij}^{*}$")
plt.savefig('./Images/Wij.png',dpi=100,bbox_inches='tight')
plt.show()
```

#### 7.c) Analysis and Comparison with ER Graphs

```python
# ER Graphs
n_nodes  = nx.number_of_nodes(G)
all_deg  = list(dict(list(nx.degree(G))).values()) 
p_ER     = np.mean(all_deg) / (n_nodes - 1)

# simulate 200 ER graphs

np.random.seed(1234)

runs     = 200
deg_ER   = np.empty((runs, n_nodes))
clust_ER = np.empty((runs,n_nodes))
asrt_ER  = np.empty(runs)
btw_ER   = np.empty((runs,n_nodes))
diam_ER  = 0
for r in range(runs):
    G_ER          = nx.erdos_renyi_graph(n_nodes,p_ER, seed=np.random)
    deg_ER[r,:]   = np.array(list(dict(nx.degree(G_ER)).values()))
    clust_ER[r,:] = np.array(list(nx.clustering(G_ER).values()))
    asrt_ER[r]    = nx.degree_assortativity_coefficient(G_ER)
    btw_ER[r,:]   = np.array(list(nx.betweenness_centrality(G_ER).values()))
    diam_ER       += nx.diameter(G_ER)
diam_ER /= runs

print("-- ER --")
print('Average Degree = ', np.mean(deg_ER[deg_ER!=0]))
print('Average Diameter = ', diam_ER)
print('Mean Clustering Coefficent = ', np.mean(clust_ER[clust_ER != 0]))
print('Mean Degree Assortativity Coefficent = ', np.mean(asrt_ER))
print('Mean Betweenness Centrality = ', np.mean(btw_ER))
```

```python
# Degrees
degrees = np.array(list(dict(nx.degree(G)).values()))
plt.hist(degrees,bins=25,range=(0,50),density=True, label='$W_{ij}^{*}$ Degrees')
plt.hist(deg_ER[deg_ER != 0].flatten(),bins=25,range=(0,50), density=True, alpha=0.5, label='ER Degrees')
plt.xlabel('Degree')
plt.ylabel('Density')
plt.legend()
plt.title("Degrees Comparison")
plt.savefig('./Images/Deg.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
# Clustering
clustering   = nx.clustering(G).values()
plt.figure()
plt.hist(clustering,bins=25,range=(0,1),density=True, label='$W_{ij}^{*}$')
plt.hist(clust_ER[clust_ER != 0].flatten(),bins=25,range=(0,1), density=True, alpha=0.5, label='ER')
plt.xlabel('Clustering Coefficents')
plt.ylabel('Density')
plt.legend()
plt.title("Clustering Coeff. Comparison")
plt.savefig('./Images/Cluster.png',dpi=100,bbox_inches='tight')
plt.show()
print('Mean Clustering Coefficent = ', np.mean(list(clustering)))
```

```python
# Diameter
diam = nx.diameter(G)
print('The diameter is:', diam)

# Assortativity
asrt = nx.degree_assortativity_coefficient(G)
print('Degree Assortativity Coefficent = ', asrt)

# Betweenness Centrality
btw  = nx.betweenness_centrality(G,weight='weighted')
print('Mean Betweenness Centrality = ', np.mean(list(btw.values())))
```
