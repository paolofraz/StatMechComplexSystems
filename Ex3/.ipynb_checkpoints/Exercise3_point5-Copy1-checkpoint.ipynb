```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import lines
```

```python
P_n=np.zeros(1)
b=1
m=0.1
K=10
```

```python
for n in range (1,91,1):
    num=1
    den=1
    for j in range (0,n,1):
        num=num*(b*(j)+m)
    for j in range (1,n+1,1):
        den=den*j**(2)
    num=num*K**(n)
    den=den*b**(n)


    P_n=np.append(P_n,num/den)

print(P_n[90]/P_n[10])    
P_0=1/(np.sum(P_n))

P_n[0]=P_0

P_n=P_n*P_0
```

```python
plt.figure(figsize=(12,8))
plt.grid()
plt.plot(P_n)
plt.axvline(10, color='grey', linestyle='--', linewidth=3)
plt.xlabel("$n$", fontsize=14)
#plt.xticks(list(plt.xticks()[0]) + [1])
plt.ylabel("Density", fontsize=14)
plt.xlim((1,40))
plt.ylim((0,0.13))
plt.xticks((1,5,10,15,20,25,30,35,40),(1,5,"$K=10$",15,20,25,30,35,40))
#plt.ylim(0,1)
#plt.title("Cumulative Plot of the Relative Specie Abundance (RSA)", fontsize=18)
plt.savefig('./Images/Point5.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
n=np.arange(0,91,1)
med=np.sum(P_n*n)
print(med)
```

```python
#Gillespie
tempo=np.zeros(1)
state=np.array([90])                #initial state

for n in range(0,2000,1):             #number of interactions
    n=state[len(state)-1]
    a1=b*n+m                        
    a2=(b*n**2)/K
    a0=a1+a2
    
    x=np.random.uniform(0,1)
    t=-np.log(x/a0)/(a0)
    tempo=np.append(tempo,tempo[len(tempo)-1]+t)
    
    y=np.random.uniform(0,1)
    if (y<a1/a0):                                       #birth
        state=np.append(state,state[len(state)-1]+1)
    else:                                               #death
        state=np.append(state,state[len(state)-1]-1)

plt.plot(tempo,state)
plt.show()
plt.hist(state,range=(0,40))
plt.show()

    


```
